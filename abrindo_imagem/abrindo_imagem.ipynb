```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread("pessoas-felizes.jpg")
```

```python
plt.imshow(imagem)
```

 # Fazendo converso para exibir corretamente a imagem

```python
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem_rgb)
```

```python
imagem_rgb.shape
```

```python
imagem_gray = cv2.cvtColor(imagem_rgb, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(imagem_gray, cmap="gray")
# para criar uma imagem em padro de cinza, basta calcular a mdia dos trs canais gerados (imagem_rgb.shape) de cada pixel
```

# ROI - Region Of Interest

```python
imagem = cv2.imread("pessoas-escritorio.jpg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem)
```

```python
imagem.shape
# R, G, B
```

```python
imagem_roi = imagem[150:250, 300:500]
```

```python
plt.imshow(imagem_roi)
```

```python
cv2.imwrite("imagem_roi.jpg", imagem_roi)
```

```python
imagem_roi_bgr = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR)
cv2.imwrite("imagem_roi.jpg", imagem_roi_bgr)
```
